```python
mw = {'CH4':16.04, 'H20':18.02, 'O2':32.00, 'CO2':44.01}
print(mw)
```

```python
mw['C8H18'] = 114.23
print(mw)
```

```python
mw['CH4']
```

```python
for species in mw.keys():
    print("The molar mass of {:<s} is {:<7.2f}".format(species, mw[species]))
```

```python
for species in sorted(mw):
    print("{:<8s} {:>7.2f}".format(species, mw[species]))
```

```python
for species in sorted(mw, key=mw.get):
    print("{:s} {:f}".format(species, mw[species]))
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.linspace(0,10)
y = np.sin(x)
z = np.cos(x)

plt.plot(x,y, 'b', x,z,'r')
plt.xlabel('Radian')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

```python
plt.plot(y,z)
plt.axis('equal')
```

```python
plt.subplot(2,1,1)
plt.plot(x,y)
plt.title('Sin(X)')

plt.subplot(2,1,2)
plt.plot(x,z)
plt.title('Cos(X)')
```
