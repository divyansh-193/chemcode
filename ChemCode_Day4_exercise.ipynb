###Generating some random data

1. import the numpy and matplotlib packages
2. use numpy to generate a 100X100 'randn' array and store in a variable 'lar_arr'
3. Get the shape of this array and store in a variable 's'
4. For the 'y' values, let's do an operation ,

    lar_arr + np.square(lar_arr) + np.ones(s)

5. Because the size of this 'y' array is 10,000 numbers, generate an array 'x' with 10,000 evenly spaced numbers starting from 0. You have many options here. Use the one you are comfortable with.
6. Now, 'unravel' the array 'y' because to plot these numbers, we need 'x' & 'y' to have the same shape and dimension.

###**Hint**: There are many numpy functions to use in this exercise. But we have covered them all in the lecture videos. If you break them down into small tasks, it will be very simple.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from numpy.random import randn
lar_arr = randn(100,100)
```

```python
s = lar_arr.shape
```

```python
y = lar_arr + np.square(lar_arr) + np.ones(s)
x = np.arange(0,10000)
```

```python
y = y.ravel()
```
